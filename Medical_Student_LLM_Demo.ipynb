```python
##Test 

# Function to process responses and calculate percentages iteratively
def process_responses(df):
    results = []
    
    # Iterate through each column except 'Person'
    for column in df.columns[1:]:
        # Initialize a counter dictionary
        counts = {}
        
        # Process the column
        for _, row in df.iterrows():
            value = row[column]
            # Check if the value is a string before processing
            if isinstance(value, str):
                responses = value.split(',') if ',' in value else [value]
                for response in responses:
                    response = response.strip()  # Clean up whitespace
                    counts[response] = counts.get(response, 0) + 1
        
        total_responses = sum(counts.values())
        
        for response, count in counts.items():
            percentage = (count / total_responses) * 100
            results.append({
                'Question': column,
                'Response': response,
                'Count': count,
                'Total Responses': total_responses,
                'Percentage': f"{percentage:.2f}%"
            })
    
    return pd.DataFrame(results)


# Process the DataFrame
test_split_result_df = process_responses(perspective_llm_use_df_clean_selected)
```

```python
def segmentation_analysis(df, demographic_columns, question_columns):
    for demographic in demographic_columns:
        print(f"Segmenting by: {demographic}")
        
        # Get unique values for the demographic column
        unique_values = df[demographic].dropna().unique()
        
        for group in unique_values:
            print(f"\nAnalyzing for {demographic} = {group}")
            
            # Filter the DataFrame for the current group
            group_df = df[df[demographic] == group]
            
            for question in question_columns:
                print(f"  Question: {question}")
                
                # Extract responses for the question
                responses = group_df[question].dropna()
                
                # Flatten multiple responses (e.g., "Chat GPT, Google Bard")
                flattened_responses = responses.str.split(',').explode().str.strip()
                
                # Count occurrences of each response
                counts = flattened_responses.value_counts()
                
                # Calculate percentages
                total_responses = len(responses)
                percentages = (counts / total_responses) * 100
                
                # Visualization
                plt.figure(figsize=(10, 6))
                plt.bar(counts.index, percentages, alpha=0.7, color=plt.cm.tab10.colors[:len(counts)])
                plt.title(f"{question} - {demographic}: {group}", fontsize=14)
                plt.ylabel("Percentage (%)", fontsize=12)
                plt.xlabel("Response", fontsize=12)
                plt.xticks(rotation=45, ha='right')
                plt.tight_layout()
                
                # Save the plot
                filename = f"segmented_{demographic}_{group}_{question[:50]}.png".replace("/", "_").replace(" ", "_")
                plt.savefig(filename)
                plt.close()
                print(f"    Saved plot as {filename}")


```

```python
def process_responses_fixed(df):
    results = []
    
    # Iterate through each column except 'Person'
    for column in df.columns[1:]:
        # Initialize a counter dictionary
        counts = {}
        total_rows_with_responses = 0  # Track rows with at least one response
        
        # Process the column
        for _, row in df.iterrows():
            value = row[column]
            # Check if the value is a string before processing
            if isinstance(value, str):
                responses = value.split(',') if ',' in value else [value]
                for response in responses:
                    response = response.strip()  # Clean up whitespace
                    counts[response] = counts.get(response, 0) + 1
                total_rows_with_responses += 1  # Increment for rows with a response
        
        # Use total_rows_with_responses instead of sum(counts.values())
        for response, count in counts.items():
            percentage = (count / total_rows_with_responses) * 100
            results.append({
                'Question': column,
                'Response': response,
                'Count': count,
                'Total Responses': total_rows_with_responses,
                'Percentage': f"{percentage:.2f}%"
            })
    
    return pd.DataFrame(results)

```

```python
def process_responses_by_multiple_demographics(df, demographic_columns):
    results = []

    # Iterate through each demographic column
    for demographic_column in demographic_columns:
        print(f"Processing responses for demographic: {demographic_column}")

        # Get unique values for the demographic column
        unique_groups = df[demographic_column].dropna().unique()

        # Iterate through each unique demographic group
        for group in unique_groups:
            print(f"  Analyzing group: {group}")
            
            # Filter the DataFrame for the current group
            group_df = df[df[demographic_column] == group]
            
            # Iterate through each question column
            for column in df.columns:
                if column in demographic_columns:  # Skip demographic columns
                    continue
                
                # Initialize a counter dictionary
                counts = {}
                total_rows_with_responses = 0  # Track rows with at least one response
                
                # Process the column
                for _, row in group_df.iterrows():
                    value = row[column]
                    if isinstance(value, str):  # Ensure the value is a string
                        responses = value.split(',') if ',' in value else [value]
                        for response in responses:
                            response = response.strip()  # Clean up whitespace
                            counts[response] = counts.get(response, 0) + 1
                        total_rows_with_responses += 1
                
                # Add results for this question and group
                for response, count in counts.items():
                    percentage = (count / total_rows_with_responses) * 100
                    results.append({
                        'Demographic Group': demographic_column,
                        'Group Value': group,
                        'Question': column,
                        'Response': response,
                        'Count': count,
                        'Total Responses': total_rows_with_responses,
                        'Percentage': f"{percentage:.2f}%"
                    })

    # Return the results as a DataFrame
    return pd.DataFrame(results)

```

```python
##VIsualizations
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

def visualize_responses_no_errors(df):
    # Iterate through unique questions in the DataFrame
    for question in df['Question'].unique():
        # Filter the DataFrame for the current question
        data = df[df['Question'] == question]
        
        # Convert the 'Percentage' column to numeric
        data['Percentage'] = data['Percentage'].str.rstrip('%').astype(float)
        
        # Assign colors to each response
        unique_responses = data['Response'].unique()
        colors = plt.cm.tab10(range(len(unique_responses)))  # Use a colormap
        
        # Plot bar chart 
        plt.figure(figsize=(10, 6))
        plt.bar(
            data['Response'],
            data['Percentage'],
            color=colors,
            alpha=0.7,
            capsize=5
        )
        plt.title(question, fontsize=14)
        plt.ylabel('Percentage (%)', fontsize=12)
        plt.xlabel('Response', fontsize=12)
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

# Call the function with your DataFrame
visualize_responses_no_errors(test_split_fixed_df)


# # Call the function with your DataFrame
# visualize_responses(test_split_fixed_df)
```

```python
def pie_chart(df):
    for question in df['Question'].unique():
        # Filter the DataFrame for the current question
        data = df[df['Question'] == question]

        # Plot pie chart
        plt.figure(figsize=(8, 8))
        plt.pie(
            data['Count'],
            labels=data['Response'],
            autopct='%1.1f%%',
            startangle=140,
            colors=plt.cm.tab10.colors[:len(data)]
        )
        plt.title(f"Response Distribution: {question}", fontsize=14)
        plt.show()

pie_chart(test_split_fixed_df)
```
